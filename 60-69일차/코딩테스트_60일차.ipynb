```python
# https://www.acmicpc.net/problem/16935 (시뮬레이션) 실버 2
```

```python
import sys
# sys.stdin.readline()
n, m, r = map(int, sys.stdin.readline().split())
t = []

for _ in range(n):
    t.append(list(map(int, sys.stdin.readline().split())))

order = list(map(int, sys.stdin.readline().split()))

def one(table):
    n_table = []
    for i in range(len(table)-1, -1, -1):
        n_table.append(table[i])
    return n_table
    
def two(table):
    n_table = [[0] * len(table[0]) for _ in range(len(table))]
    for i in range(len(table)):
        for j in range(len(table[0])):
            n_table[i][j] = table[i][len(table[0])-j-1]
    return n_table
    
def three(table):
    n_table = [[] for _ in range(len(table[0]))]
    for i in range(len(table[0])):
        for j in range(len(table)):
            n_table[i].append(table[len(table)-1-j][i])
    return n_table
    
def four(table):
    n_table = [[] for _ in range(len(table[0]))]
    for i in range(len(table[0])):
        for j in range(len(table)):
            n_table[i].append(table[j][len(table[0])-i-1])
    return n_table

def five(table):
    n_table = [[0] * len(table[0]) for _ in range(len(table))]
    for i in range(len(table)):
        for j in range(len(table[0])):
            if 0 <= i < len(table)//2:
                if 0 <= j < len(table[0])//2:
                    n_table[i][j] = table[len(table)//2+i][j]  # 1번째 자리
                else:
                    n_table[i][j] = table[i][j-len(table[0])//2] # 2번째 자리
            else:
                if 0 <= j < len(table[0])//2:
                    n_table[i][j] = table[i][len(table[0])//2+j] # 4번째 자리 
                else:
                    n_table[i][j] = table[i-len(table)//2][j]# 3번째 자리
    return n_table
                    
            
    
def six(table):
    n_table = [[0] * len(table[0]) for _ in range(len(table))]
    for i in range(len(table)):
        for j in range(len(table[0])):
            if 0 <= i < len(table)//2:
                if 0 <= j < len(table[0])//2:
                    n_table[i][j] = table[i][len(table[0])//2+j]  # 1번째 자리
                else:
                    n_table[i][j] = table[len(table)//2+i][j] # 2번째 자리
            else:
                if 0 <= j < len(table[0])//2:
                    n_table[i][j] = table[i-len(table)//2][j] # 4번째 자리 
                else:
                    n_table[i][j] = table[i][j-len(table[0])//2]# 3번째 자리
    return n_table
    
for i in order:
    if i == 1:
        t = one(t)
    elif i == 2:
        t = two(t)
    elif i == 3:
        t = three(t)
    elif i == 4:
        t = four(t)
    elif i == 5:
        t = five(t)
    else:
        t = six(t)
        
for i in range(len(t)):
    for j in range(len(t[0])):
        print(t[i][j], end = " ")
    print()   
```
