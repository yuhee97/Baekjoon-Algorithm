##### 구현 문제 1: 그룹 단어 체커  https://www.acmicpc.net/problem/1316


```python
import sys

N = int(sys.stdin.readline())
count = 0

for i in range(N):
    str = ''
    word = sys.stdin.readline().strip()
    for j in range(len(word)):
        if word[j-1] == word[j]:
            str = str + word[j]
            continue
        if word[j] in str:
            break
        str = str + word[j]
    if len(str) == len(word):
        count += 1
print(count)

# sorted된 상태랑 비교하면 더 짧게 구현 가능하다 **
```

##### 구현 문제 2: 덩치 https://www.acmicpc.net/problem/7568


```python
import sys

N = int(input())
rank = [1]*N
people = []

for i in range(N):
    kg, cm = map(int, input().split())
    people.append([kg, cm])
    
for i in range(N):
    for j in range(N):
        if people[i][0] > people[j][0] and people[i][1] > people[j][1]:
            rank[j] += 1
        
print(* rank)
```

##### 구현 문제 3: 로봇 청소기 https://www.acmicpc.net/problem/14503


```python
import sys

N, M = map(int, sys.stdin.readline().split())
r, c, d = map(int, sys.stdin.readline().split())
lis = []; count = 0
dx, dy = (-1, 0, 1, 0), (0, 1, 0, -1) # 북쪽(0), 동쪽(1), 남쪽(2), 서쪽(3)

for _ in range(N):
    lis.append(list(map(int, sys.stdin.readline().split())))
    
if lis[r][c] == 0:
    lis[r][c] = 2
    count = 1

while(True):
    check = False
    for i in range(4):
        nd = (d + 3) % 4 # 회전
        x = r + dx[nd] 
        y = c + dy[nd]
        d = nd
        if lis[x][y] == 0:
            check = True
            lis[x][y] = 2
            count += 1
            r, c = x, y
            break
    
    if not check:
        if lis[r - dx[d]][c - dy[d]] == 1:
            break
        else:
            r, c = r - dx[d], c - dy[d]
        
print(count)
```
