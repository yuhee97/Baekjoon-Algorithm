##### 문제 1: 바이러스(실버3)
https://www.acmicpc.net/problem/2606

```python
import sys
from collections import deque

N = int(sys.stdin.readline()) 
M = int(sys.stdin.readline()) 
graph = [[] for i in range(N+1)]
visited = [0 for i in range(N+1)]

for i in range(M):
    a, b = map(int, input().split())
    graph[a].append(b)
    graph[b].append(a)

def bfs(v):
    queue = deque()
    queue.append(v)
    visited[v] = 1
    while queue:
        v_new = queue.popleft()
        for i in graph[v_new]:
            if visited[i] == 0:
                queue.append(i)
                visited[i] = 1
    return sum(visited)
    
print(bfs(1) - 1)  
```

##### 문제 2: 숨바꼭질(실버1)
https://www.acmicpc.net/problem/1697

```python
import sys
from collections import deque

N, K = map(int,input().split()); MAX = 100001
visited = [0 for _ in range(MAX)]
graph = [0 for _ in range(MAX)]

queue = deque()
queue.append(N)
visited[N] = 1

while queue:
    x = queue.popleft()
    if x == K:
        print(graph[x])
        break   

    if 0 <= x-1 < MAX and visited[x-1] == 0:
        queue.append(x-1)
        visited[x-1] = 1
        graph[x-1] = graph[x] + 1
    if 0 <= x+1 < MAX and visited[x+1] == 0:
        queue.append(x+1)
        visited[x+1] = 1
        graph[x+1] = graph[x] + 1
    if 0 <= x*2 < MAX and visited[x*2] == 0:
        queue.append(x*2)
        visited[x*2] = 1
        graph[x*2] = graph[x] + 1
```

##### 문제 3: 유기농 배추
https://www.acmicpc.net/problem/1012

```python
import sys
from collections import deque

T = int(sys.stdin.readline())

for _ in range(T):
    M, N, K = map(int, sys.stdin.readline().split())
    graph = [[0] * N for _ in range(M)]
    visited = [[0] * N for _ in range(M)]

    for _ in range(K):
        a, b = map(int, sys.stdin.readline().split())
        graph[a][b] = 1

    dx = [-1, 0, 1, 0]
    dy = [0, -1, 0, 1]
    
    queue = deque()
    count = False
    c = 0
    for i in range(M):
        for j in range(N):
            if visited[i][j] == 0 and graph[i][j] == 1:
                queue.append((i, j))
                visited[i][j] = 1
                c += 1; 
                while queue:
                    x, y = queue.popleft()
                    for k in range(4):
                        nx = x + dx[k]
                        ny = y + dy[k]
                        if 0<= nx < M and 0<= ny < N and visited[nx][ny] == 0 and graph[nx][ny] == 1:
                            queue.append((nx, ny))
                            visited[nx][ny] = 1
                            count = True
    print(c)
```
